## Investigation Overview

In this investigation, I want to discover the correlations present between features in the data. The main focus was on the year, sales, genre, and gender variables. 


## Dataset Overview

The Video Game Sales dataset collected from Kaggle website. The dataset consistes of information about more than 10000 Video Games, it contains 11 columns and 16291 rows. Including, name, year, genre, publisher, and global sales. The dataset contains both qualitative and quantitative types of data.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load the dataset
df = pd.read_csv('vgsales.csv')
```

## Number of Video Games Released

This figure shows the number of video games released from 1980 to 2020. Which helps to track the number of video games that have been released over the years.

```python
# display a histogram 
plt.figure(figsize=[10, 7])
plt.title("Number of Video Games Released", fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.hist(data = df, x = 'Year', bins = 20);
```

## The Platform With The Highest Releases

This figure shows the number of video games released in each platform. Which helps to distinguish the platforms with higher releases.

```python
# display the number of video games released by Platforms
plt.figure(figsize=[10, 7])
base_color = sb.color_palette()[0]
sb.countplot(data=df, x='Platform', color=base_color);
plt.title("Number of Video Games Released", fontsize=18)
plt.xlabel('Platform', fontsize=14)
plt.ylabel("Count", fontsize=14)
# Use xticks to rotate the category labels
plt.xticks(rotation=90);
```

## The Genre With Higher Releases

This figure shows the number of video games released in each genre. Which helps to distinguish the genres with highest releases.

```python
# display the number of video games released by Platforms
plt.figure(figsize=[10, 7])
sb.countplot(data=df, x='Genre', color=base_color);
plt.title("Number of Video Games Released in Each Genre", fontsize=18)
plt.xlabel('Genre', fontsize=14)
plt.ylabel("Count", fontsize=14)
# Use xticks to rotate the category labels
plt.xticks(rotation=90);
```

## Number of Videos Released in Each Genre Among All the Years
The figure shows what's the number of video games released, it will help to distinguish the genres with the highest releases as well as the higher release years.

```python
sb.boxplot(data=df, x='Genre', y='Year', color=base_color)
plt.title("Number of Videos Released in Each Genre Among All the Years", fontsize=18)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Year', fontsize=14)
# Use xticks to rotate the category labels
plt.xticks(rotation=90);
```
