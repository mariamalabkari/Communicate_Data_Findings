# Univariate Exploration
In this section, I will start to do visualization for the individual variables and look at each different variable at a time.

#### What is the number of video games released among all the years?

```python
base_color = sb.color_palette()[0]
```

```python
# display the number of video games released by year
plt.figure(figsize=[10, 7])
sb.countplot(data=df, x='Year', color=base_color);
plt.title("Number of Video Games Released", fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel("Count", fontsize=14)
# Use xticks to rotate the category labels
plt.xticks(rotation=90);
```

> The above figure shows the number of video games released from 1980 to 2020. It is clear that the number of video games started to increase year by year. However, in 2009 the number of videos games released started to decrease.

```python
# display the number of video games released by year
sb.distplot(df.Year, bins = 100,  kde=False);
plt.title("Number of Video Games Released", fontsize=18)
plt.ylabel('Year', fontsize=14);
plt.xlabel("Count", fontsize=14);
```

> The figure above also shows the number of video games released from 1980 to 2020 but with using the dist plot. It is clear that the number of video games started to increase year by year. However, in 2009 the number of videos games released started to decrease.

```python
# display a histogram 
plt.figure(figsize=[10, 7])
plt.title("Number of Video Games Released", fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.hist(data = df, x = 'Year', bins = 20);
```

> The histogram above also shows the number of video games released from 1980 to 2020. It is clear that the number of video games started to increase year by year. However, in 2009 the number of videos games released started to decrease.

#### What is the relationship between all variables?

```python
numeric_vars = ['Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
categoric_vars = ['Platform', 'Genre', 'Publisher']
```

# Multivariate Exploration
> In this section I will create plots of three or more variables to investigate the data even further. 

#### Dose the year, genre, and global sales have an impact on each other?

```python
g = sb.PairGrid(data = df, vars = ['Year', 'Genre', 'Global_Sales'])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

> This figure shows the relationship between year, genre, and global sales. It's clear that global sales increase with the increase in the year.

#### Dose the year, global sales, and other sales have an impact on each other?

```python
g = sb.PairGrid(data = df, vars = ['Year', 'Other_Sales', 'Global_Sales'])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

> This figure shows the relationship between the year, global sales, and other sales. It's clear that global sales and the other sales both increase with the increase in the year.

#### What is the relationship between sales in Europe, Japan, and North America?

```python
plt.scatter(data = df, x = 'EU_Sales', y = 'JP_Sales', s = 'NA_Sales')

# dummy series for adding legend
sizes = [20, 40, 50]
base_color = sb.color_palette()[0]
legend_obj = []
for s in sizes:
    legend_obj.append(plt.scatter([], [], s = s, color = base_color))
plt.legend(legend_obj, ['Europe','Japan','North America'])
plt.title("Sales Among Europe, Japan, and North America", fontsize=18)
plt.xlabel('Sales (millions)', fontsize=14);
plt.ylabel('Count', fontsize=14);
```

> The above scatter plot shows the relationship between sales in Europe, Japan, and North America. It is clear from the figure that the sales were around 0 and 5 million.
